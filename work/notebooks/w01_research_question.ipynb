## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

What decision does your work improve?
Which content pages a review team should prioritize first for refresh, expansion, or pruning — instead of reviewing pages in an arbitrary order or relying purely on a fixed rule.
Who acts on it?
A content reviewer or SEO strategist, who takes the ranked list and manually decides what to do with each top-ranked page (refresh it, expand it, leave it, or flag it for removal).
What does a wrong recommendation cost?

If a page is flagged as high-priority but isn't actually a problem, the team spends limited review time on it for nothing.
If a genuinely declining, high-traffic page is missed (not flagged), it keeps losing visibility silently — and that's the more expensive mistake, since traffic loss compounds the longer it goes unnoticed.
